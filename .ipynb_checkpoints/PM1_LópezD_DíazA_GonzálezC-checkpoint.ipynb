## 1. 4 Solución del problema de optimización

```python
import numpy as np
import scipy.optimize as opt
```

```python
c=np.array([0,1,1,1,1,
            -1.5648,0,0,0,0,
           -0.1856,0,0,0,0,
           -0.6361,0,0,0,0,
           -0.01011,0,0,0,0])

A_ub=np.array([[0,-1,-1,-1,-1,
               1.5648,0,0,0,0,
               0.1856,0,0,0,0,
               0.6361,0,0,0,0,
               0.01011,0,0,0,0],
              [0,-0.6390,0,0,0,
              1,0,1,1,1,
              0,-0.1186,0,0,0,
              0,-0.4063,0,0,0,
              0,-0.00645,0,0,0],
              [0,0,-5.3712,0,0,
              0,0,-8.4304,0,0,
              1,1,0,1,1,
              0,0,-3.4233,0,0,
              0,0,-0.05431,0,0],
              [0,0,0,-1.5712,0,
              0,0,0,-2.4590,0,
              0,0,0,-0.2921,0,
              1,1,1,0,1,
              0,0,0,-0.01588,0],
              [0,0,0,0,-98.8901,
              0,0,0,0,-154.7733,
              0,0,0,0,-18.4122,
              0,0,0,0,-62.94,
              1,1,1,1,0]])

B_ub=np.array([1,0,0,0,0])

xi_bounds=(0,6000)
```

```python
sol=opt.linprog(c,A_ub,B_ub,None,None,xi_bounds)
sol
```
